```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ['AAPL', 'GOOGL']

train_start_date = "2023-10-09"
train_end_date = "2023-10-14"

time_frame = '1m'
```

```python
df = None

for ticker in tickers:
    data = yf.download(tickers=ticker,
                    start=train_start_date,
                    end=train_end_date,
                    interval=time_frame
                    )
    # set datetime to a column
    data.reset_index(inplace=True)

    # all columns to lowercase
    data.columns = data.columns.str.lower()

    #insert tic column
    data.insert(0, 'tic', [ticker] * data.shape[0])

    # get sma
    close_5_sma = data['close'].rolling(window=5).mean()
    close_20_sma = data['close'].rolling(window=30).mean()
    close_60_sma = data['close'].rolling(window=60).mean()

    # get bollinger band
    std = data['close'].rolling(window=20).std()
    boll_ub = close_20_sma + 2 * std
    bool_lb = close_20_sma - 2 * std

    # insert indicators
    data.insert(8, 'close_5_sam', close_5_sma)
    data.insert(8, 'close_20_sam', close_20_sma)
    data.insert(8, 'close_60_sam', close_60_sma)
    data.insert(8, 'boll_ub', boll_ub)
    data.insert(8, 'boll_lb', bool_lb)

    # cut out first 60 data
    data = data[:][60:]
    
    if df is None:
        df = data
    else:
        df = pd.concat([df, data], axis=0)
```
